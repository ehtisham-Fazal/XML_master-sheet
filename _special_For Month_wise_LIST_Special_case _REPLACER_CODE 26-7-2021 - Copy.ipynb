```python
#for finding first charater:
def find(s, ch):
    return [i for i, ltr in enumerate(s) if ltr == ch]

#for replaceing %
def replace_precentage():
    print("______starting______")
    counter=0
    curr = 0
    for row in sheet.iter_rows(min_row=2, max_col=55, max_row=720):
         for cell in row:
            #enter logic here
            str1=str(cell.value)
            indexes = find(str1,"%")
            if len(indexes)>0:
                print("Product ID : "+ str(row[0].value))
                #print(get_column_letter(cell.column))
                my_letter=get_column_letter(cell.column)
                for key, value in activity_title_dic.items():
                    if key == my_letter:
                        print(value)
                        print("")
            
                str_indx = 0
                fin_str = str1
                   
            for r in range(0,len(indexes)):
                #print(r)
                rem_indexes = find(fin_str,"%")
                #print(rem_indexes)
                tempstart = fin_str[curr:rem_indexes[0]]
                #print(tempstart)
                char = fin_str[rem_indexes[0]+1:rem_indexes[0]+3]
                #print(char)
                tempend = fin_str[rem_indexes[0]+3:]
                #print(tempend)
                fin_str = tempstart + '\\x' + char + '\\x' +tempend
                #print(fin_str)
                #print()

                #print("FINAL STRING")
                #print(fin_str)
                cell.value=fin_str
                if str1!=cell.value:
                 counter=counter+1

#             str_indx = 0
#             new_str = str1
#             fin_str = ""
#             for i in indexes:
#                 rpc_string = str1[i:i+3]
#                 temp1 = new_str[str_indx:i] + "\\x" + rpc_string[1:] + "\\x"
#                 fin_str = fin_str + temp1
#                 str_indx = i+3
#                 if len(indexes)!=0:
#                     print("indexes : ", indexes)
#                     print(cell.value)
#                     cell.value = fin_str + cell.value[indexes[-1]+3:]
                 
#                 if str1!=cell.value:
#                     counter=counter+1
    print("% replace to/x /x; no of cell replace : "+str(counter))
    print("____!!___!!_ended__!!__!!___")     
    
#for replaceing &amp;   
def replace_andapm():
    print("______starting_____")
    counter=0
    for row in sheet.iter_rows(min_row=2, max_col=55, max_row=720):
         for cell in row:
            #enter logic here
            str1=str(cell.value)
            indexes = find(str1,"<")
            if len(indexes)>0:
              if "&amp;" not in str1 and "&" in str1:
                new_str = str1.replace(r"&", "&amp;")
                print("Product ID : "+ str(row[0].value))
                #print(get_column_letter(cell.column))
                my_letter=get_column_letter(cell.column)
                for key, value in activity_title_dic.items():
                    if key == my_letter:
                        print(value)
                        print("")
                if str1!=new_str:
                    counter=counter+1
                cell.value=new_str
                #print(new_str)
    print("& replace to &amp; no of cell replace : "+str(counter))
    print("____!!___!!_ended__!!__!!___")   
    
    
#for replaceing /x/x    
def replace_backslashx():
    print("______starting_____")
    counter=0
    for row in sheet.iter_rows(min_row=2, max_col=55, max_row=720):
         for cell in row:
            #enter logic here
            str1=str(cell.value)
            indexes = find(str1,"<")
            if len(indexes)>0:
              if r"\x\x"  in str1:
                new_str = str1.replace(r"\x\x", " ")
                print("Product ID : "+ str(row[0].value))
                #print(get_column_letter(cell.column))
                my_letter=get_column_letter(cell.column)
                for key, value in activity_title_dic.items():
                    if key == my_letter:
                        print(value)
                        print("")
                if str1!=new_str:
                    counter=counter+1
                cell.value=new_str
                #print(new_str)
    print("/x/x replace to space no of cell replace : "+str(counter))
    print("____!!___!!_ended__!!__!!___")
```
